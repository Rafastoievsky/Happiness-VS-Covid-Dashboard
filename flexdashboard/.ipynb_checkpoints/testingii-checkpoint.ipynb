```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
complete = pd.read_csv('../data/completedata.csv')
complete.head()
```

```python
x =complete.loc[:,'Confirmed':'dystopia_residual'].values 

y = complete.loc[:,['continent']].values
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
```

```python
pca_df = pd.DataFrame(principalComponents, columns=['PC1', 'PC2', 'PC3'])
pca_df['continent'] = y
pca_df.head()
```

## Col 2 (tabs)

```python
sns.set_context("notebook", font_scale=1.1)
sns.set_style("ticks")

sns.lmplot(x='PC1',
           y='PC2',
          data=pca_df,
          fit_reg=False,
          legend=True,
          height=9,
          hue='continent',
          scatter_kws={"s":200, "alpha":0.3})
plt.title('PCA Result Digits', weight='bold').set_fontsize('14')
plt.xlabel('Prim Comp 1', weight='bold').set_fontsize('10')
plt.ylabel('Prim Comp 2', weight='bold').set_fontsize('10')
```
